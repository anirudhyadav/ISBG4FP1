```python
import ast
import json
import pandas as pd
```

```python
# Your existing DataFrame
jdpreparator = pd.read_csv('jd_data_for_clustering.csv', sep=',')

# Get user input for the profile
user_profile = input("Enter the profile for which you want unique tech stacks: ")


# Create an empty list to store dictionaries
data_list = []

# Iterate through rows of the original DataFrame
for index, row in jdpreparator.iterrows():
    profile = row['profile']
    tech_stack = row['tech_stack']

    try:
        # Remove square brackets and split the string into a list
        input_skills_list = [item.strip() for item in tech_stack.strip('[]').split(',')]


        # Append each item in the tech_stack list to the list
        for item in input_skills_list:
            data_list.append({'profile': profile, 'Tech_Stack_Item': item})
    except (ValueError, SyntaxError):
        print(f"Error decoding list in row {index}")

# Create a new DataFrame from the list of dictionaries
output_df = pd.DataFrame(data_list)


# Filter the DataFrame based on user input
filtered_df = output_df[output_df['profile'] == user_profile]
# print(filtered_df)

# Check if the profile exists in the DataFrame
if not filtered_df.empty:
    # Flatten the nested lists and get unique values
    unique_tech_stack = list( filtered_df['Tech_Stack_Item'])
    print(f"Unique tech stacks for profile '{user_profile}' in job market:\n{unique_tech_stack}")
else:
    print(f"No data found for profile '{user_profile}'.")
```
